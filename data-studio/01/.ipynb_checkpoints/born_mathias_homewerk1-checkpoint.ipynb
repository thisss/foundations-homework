## PART TWO

Using the NBA data, make a chart that highlights two specific players that make either a lot or a little money vs. how well they perform. Add their names and a little information about them to the chart. Play around with fonts and colors, make that chart look great!

```python
# Import the performance data
performance_df = pd.read_excel("../../../data/NBA stats.xlsx", sheetname="Player performance")
performance_df.head()
```

```python
# Import the salary data
salary_df = pd.read_excel("../../../data/NBA stats.xlsx", sheetname="2016-2017 Salaries")
salary_df.head()
```

```python
# Merge the two together
df = performance_df.merge(salary_df, left_on='Player', right_on='NAME')
df.head()
```

```python
df[['Player', 'SALARY', 'PTS']].sort_values('SALARY', ascending=False).head()
```

```python
df[['Player', 'SALARY', 'PTS']].sort_values('PTS', ascending=False).head()
```

```python
df[['Player', 'SALARY', 'PTS']].sort_values('PTS', ascending=True).head()
```

```python
df['PTS_Salary'] = df['SALARY'] / df['PTS']
df[['Player', 'SALARY', 'PTS', 'PTS_Salary']].sort_values('PTS_Salary', ascending=False).head()
```

```python
df[df['SALARY'] > 15000000][['Player', 'SALARY', 'PTS', 'PTS_Salary', 'POS']].sort_values('PTS_Salary', ascending=False)
```

```python
fig, ax = plt.subplots(figsize=(13,8))
for category, group in df.groupby('POS'):
    ax.plot(group['PTS'], group['SALARY'], marker='o', linestyle='', label='POS', markeredgewidth=0)
ax.plot(30.1, 12112359, marker='o', markeredgewidth=18, linestyle='', label='Stephen Curry')
ax.plot(4.3,17000000, label='Joakim Noah', marker='o', markeredgewidth=18, linestyle='')
# ax.plot(15.3, 26540100, marker= 'o', markeredgewidth=18, linestyle='', label='Mike Conley')
#ax.plot(0.5, 1589640, label='Bruno Caboclo', marker='o', markeredgewidth=18)
ax.legend(numpoints = 1)
ax.set_xlim([0, 35])
ax.set_ylim([0, 30000000])
ax.set_ylabel("Earnings in Million Dollar")
ax.set_xlabel("PTS")
ax.set_title("Stephen Curry scores well, but is poorly paid")
plt.savefig('nba-points-salary.svg', transparency=True)
```

## PART THREE

Using the scrabble data, reproduce the ["Nigel Richards's Performance at Nationals"](http://fivethirtyeight.com/features/what-makes-nigel-richards-the-best-scrabble-player-on-earth/) chart from FiveThirtyEight (minus the UFO bits). Build the basic chart from this dataframe:

Hot tip: You might want to start with matplotlib using the fivethirtyeight style. Check [this blog post](https://luanneseymour.wordpress.com/2012/10/30/creating-arrows-and-arrowheads-in-illustrator-cs6/) for info on how to make arrows.

```python
points_df  = pd.read_csv("../../../data/scrabble-point-spread.csv")
tournament_df = pd.read_csv("../../../data/scrabble-tournament.csv")
df = tournament_df.merge(points_df, left_on='Player ID', right_on='player_id').fillna(0)
df.head()
```

```python
plt.style.use('fivethirtyeight')
df[['Name','points_against','points_for']].plot(kind='scatter', x='points_for', y='points_against')
```

```python
df['Perf Rating'].head(5)
```

```python
ig, ax = plt.subplots(figsize=(10,10))
for category, group in df.groupby('division'):
    ax.plot(group['points_for'], group['points_against'], marker='o', linestyle='', label=category, markeredgewidth=0)
ax.legend(numpoints = 1)
ax.set_xlim([300, 460])
ax.set_ylim([300, 460])
ax.set_title("Richard rules them all")

#plt.savefig('scrabble.svg', transparency=True)
```
